# Предобработка данных

```python
import nltk 
```

```python
nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('russian')

print(len(stop_words))
```

```python
word_tokenizer = nltk.WordPunctTokenizer()
```

```python
import re
regex = re.compile(r'[А-Яа-яёЁ]+')

def words_only(text, regex=regex):
    try:
        return " ".join(regex.findall(text)).lower()
    except:
        return ""

def process_data(data):
    texts = []

    for item in tqdm(data):
        print(item)

        text_lower = words_only(item)
        tokens     = word_tokenizer.tokenize(text_lower)

        tokens = [word for word in tokens if (word not in stop_words and not word.isnumeric() and len(word) > 1)]
        print(tokens)
        print("\n")
        texts.append(tokens)

    return texts
```

```python
y = data['rating']
texts = process_data(data['text'])
```

```python
print("Tokens: ", texts[0])
print("Y: ", y[0])
```

# Нормализация слов

```python
from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer("russian")

i = 0
for aword in texts[i][:10]:
    aword_stem = stemmer.stem(aword)
    print("Before: %s, After: %s" % (aword, aword_stem))
```

```python
import pymorphy2

morph = pymorphy2.MorphAnalyzer()

i = 0
for aword in texts[i][:10]:
    aword_norm = morph.parse(aword)[0].normal_form
    print("Before: %s\tAfter: %s" % (aword, aword_norm))
```

```python
for i in range(len(texts)):
   text_lemmatized = [morph.parse(x)[0].normal_form for x in texts[i]]
   texts[i] = ' '.join(text_lemmatized)
```

```python
i = 0
print("Y: ",   y[i])
print("Text: ",  texts[i])
```

# Моделирование & Векторные представления

```python
from sklearn.model_selection import train_test_split

train_texts, test_texts, train_y, test_y = train_test_split(texts, y, test_size=0.2, random_state=42)
```

# Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features = 200)
vectorizer.fit(train_texts)

# Топ слов
vectorizer.get_feature_names_out()[:20]
```

```python
train_X = vectorizer.transform(train_texts)
train_X.todense()[:2]
```

```python
test_X  = vectorizer.transform(test_texts)
```

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor(n_estimators = 500, max_depth = 10)
reg = reg.fit(train_X, train_y)
pred = reg.predict(test_X)
```

```python
print('Предсказанный rating: ', pred[0:20], ".....")
print('Истинный rating: ', test_y[0:20], ".....")
```

```python
print("Test MSE = %.4f" % mean_squared_error(test_y, pred))
print("Test MAE = %.4f" % mean_absolute_error(test_y, pred))
```

```python
for i in range(10):
    print('Истинный rating:',y[i])
    print('Предсказанный rating:',pred[i])
    print('Текст: ', train_texts[i])
    print('\n')
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features = 200, norm = None) 
vectorizer.fit(train_texts)

vectorizer.get_feature_names_out()[:20]
```

```python
train_X = vectorizer.fit_transform(train_texts)
test_X  = vectorizer.transform(test_texts)
```

```python
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor(n_estimators = 500, max_depth = 10)
reg = reg.fit(train_X, train_y)
pred = reg.predict(test_X)
```

```python
print('Предсказанный rating: ', pred[0:20], ".....")
print('Истинный rating: ', test_y[0:20], ".....")
```

```python
print("Test MSE = %.4f" % mean_squared_error(test_y, pred))
print("Test MAE = %.4f" % mean_absolute_error(test_y, pred))
```

```python
for i in range(100):
    print('Истинный rating:',y[i])
    print('Предсказанный rating:',pred[i])
    print('Текст: ', train_texts[i])
    print('\n')
```
